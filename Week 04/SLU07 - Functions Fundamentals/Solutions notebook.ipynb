Let's practice some functions! The important part here is that you can take some plain old english and translate it into a function.

```python
# Don't mind this here, just some imports for the exercises - MAKE SURE YOU RUN THIS CELL
import inspect
from io import StringIO
import sys

class Capturing(list):
    def __enter__(self):
        self._stdout = sys.stdout
        sys.stdout = self._stringio = StringIO()
        return self
    def __exit__(self, *args):
        self.extend(self._stringio.getvalue().splitlines())
        del self._stringio    # free up some memory
        sys.stdout = self._stdout
```

### Exercise 1

Let's start with an easy one!

Implement a function that:
- takes no argument
- returns nothing
- is called "print_hello"
- prints the string "Hello"

If you are wondering what this might be useful for - imagine instead of "Hello" you wanted to print something long and nasty to type ;)

```python
# def print_hello():

### BEGIN SOLUTION
def print_hello():
    print("Hello")
### END SOLUTION
```

```python
source = inspect.getsource(print_hello)
sig = inspect.signature(print_hello)

with Capturing() as output:
    retval = print_hello()

assert retval == None
assert output[0] == "Hello"

assert 'print' in source
assert 'return' not in source
assert len(sig.parameters) == 0

print("Well done, a start was made")
```

### Exercise 4

Now for some special functions points! You need to think about this one and what you have learned up until now.

Try to implement the previous `fun_multiplying_numbers` function but now with a twist:
- Is named `fun_multiplying_numbers_with_a_twist`
- Takes two parameters
- **Multiplies the arguments without using the * operator!**
- Returns the value of the multiplication

Tip: If you remember the good old days of math, multiplication is just a sequence of adding operations in a loop and you already learned how to do those!

```python
# def fun_multiplying_numbers_with_a_twist(a,b):
    # result = 0
    # (loop)
    # return result

### BEGIN SOLUTION
def fun_multiplying_numbers_with_a_twist(a, b):
    result = 0
    for i in range(0, a):
        result += b
    return result
### END SOLUTION
```

```python
source = inspect.signature(fun_multiplying_numbers_with_a_twist)

assert fun_multiplying_numbers_with_a_twist(1, 2) == 2
assert fun_multiplying_numbers_with_a_twist(4, 5) == 20
assert fun_multiplying_numbers_with_a_twist(100, 200) == 20000
assert '*' not in inspect.getsource(fun_multiplying_numbers_with_a_twist), "Don't use the multiplication operator! ðŸ˜‰"
assert '/' not in inspect.getsource(fun_multiplying_numbers_with_a_twist), "Dividing by the inverse is the same as using the multiplication operator! ðŸ˜‰"
assert len(source.parameters) == 2
print ("All asserts passed!")
```

### Exercise 7

Implement a function that

- Is named `two_step`
- Takes 4 parameters
- Multiplies the first 2 together and stores the output in a variable called `step1`
- Divides 3rd parameter by the 4th parameter and stores the output in a variable called `step2`
- Adds the variables `step1` and `step2` together, calling it `result`
- If `result` is greater than 20, subtract 10 to it
- Returns the final result

```python
#def two_step(a,b,c,d):
### BEGIN SOLUTION
def two_step(a, b, c, d):
    step1 = a * b
    step2 = c / d

    result = step1 + step2
    if result > 20:
        result -= 10

    return result
### END SOLUTION
```

```python
source = inspect.getsource(two_step)

sig = inspect.signature(two_step)

assert two_step(2, 2, 4, 2) == 6.0, "The final result is not matching"
assert two_step(2, 3, 100, 5) == 16.0, "Are you subtracting correctly? ðŸ¤”"
assert '*' in source
assert '/' in source
assert 'step1' in source
assert 'step2' in source
assert 'result' in source
assert len(sig.parameters) == 4
print("Good!")
```

### Exercise 10a

Write a function that

1. Is called `prepare_list`
1. Takes 1 argument, a list
1. checks if all elements in the lists are real numbers
1. removes the ones that are not real numbers, without changing the order
1. Returns the list with only numbers, if no numbers were in the list, it returns an empty list

```python
# l is a list of elements for example [1,2,"a",4, 5.4]
#def prepare_list(l):
### BEGIN SOLUTION
def prepare_list(l):
    return [i for i in l if type(i) in (int,float)]
### END SOLUTION
```

```python
sig = inspect.signature(prepare_list)
assert len(sig.parameters) == 1
assert prepare_list([1,2,3]) == [1,2,3]
assert prepare_list([5,"a",1,(3,5)]) == [5,1]
assert prepare_list([[1,2,3],"a",3,'test']) == [3]
assert prepare_list([{'k':1,5:'e'}]) == []
print("There you go! almost there...")
```

### Exercise 10b

And now for your last challenge...  write a function called `list_calc` that upon receiving a list sums up all real numbers in it. Then it multiplies that result by the last list element that is a real number. In case there are no real numbers in the list return 0. Feel free to use what you created in 10a. 

```python
### BEGIN SOLUTION
def list_calc(l):
    cleaned_list=prepare_list(l)
    return_value=0
    if cleaned_list != []:
        sum_value=0
        for i in cleaned_list:
            sum_value+=i
        return_value=sum_value*cleaned_list[-1]
    return return_value

### END SOLUTION
```

```python
sig = inspect.signature(prepare_list)
assert len(sig.parameters) == 1
assert list_calc([1,2,-3]) == 0
assert list_calc([1,2,2]) == 10, "don't forget to add the last element before multiplying by it"
assert list_calc([5,"a",1,(3,5)]) == 6
assert list_calc([[1,2,3],"a",3,'test']) == 9, "if only one real number is in the cell, it is also the last element"
assert list_calc([{'k':1,5:'e'}]) == 0, "return 0 if no values are real numbers"
assert list_calc([0.5,0.2,10,-2]) == -17.4, "floats are also real numbers!"
print("Well done! I am proud of you!")
```
